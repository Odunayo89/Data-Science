```python
from pydeseq2.dds import DeseqDataSet
from pydeseq2.ds import DeseqStats

import pandas as pd
```

```python
import os
new_directory = r'C:\Users\ajibo\Desktop'
os.chdir(new_directory)
```

```python
counts = pd.read_csv('Livercorn-flaxseed.csv')
counts
```

```python
counts = counts.set_index('Geneid')
counts
```

```python
counts = counts[counts.sum(axis = 1) > 0]
counts
```

```python
counts = counts.T
```

```python
counts
```

```python
metadata = pd.DataFrame(zip(counts.index, ['LIC','LIC','LIC','LIC','LIC', 'LIF', 'LIF', 'LIF', 'LIF', 'LIF']), 
                        columns = ['Sample', 'Condition'])
```

```python
metadata = metadata.set_index('Sample')
metadata
```

```python
dds = DeseqDataSet(counts=counts,
            clinical=metadata,
            design_factors="Condition")

#Design_factors=["batch", "condition"] = ~ batch + condition
```

```python
dds.deseq2()
```

```python
stat_res = DeseqStats(dds, n_cpus=8, contrast = ('Condition', 'LIF','LIC'))
```

```python
stat_res.summary()
```

```python
res = stat_res.results_df
```

```python
res
```

```python
res = res[res.baseMean >= 10]
```

```python
res
```
